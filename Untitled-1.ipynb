```python
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min
def kmeans_and_relabel(data, n_clusters_range, n_reps):
    best_labels = None
    best_inertia = np.inf
    best_kmeans = None
    
    for n_clusters in n_clusters_range:
        print(f"Performing k-means clustering with {n_clusters} clusters...")
        
        # Perform k-means clustering with n_reps repetitions
        kmeans = KMeans(n_clusters=n_clusters, n_init=n_reps, random_state=1234)
        kmeans.fit(data)
        
        # Evaluate clustering quality
        inertia = kmeans.inertia_
        if inertia < best_inertia:
            best_inertia = inertia
            best_labels = kmeans.labels_
            best_kmeans = kmeans
    
    # Relabel clusters: finding the closest cluster centers
    relabelled_labels = relabel_clusters(best_kmeans.labels_, best_kmeans.cluster_centers_, data)
    return best_kmeans, relabelled_labels

def relabel_clusters(labels, cluster_centers, data):
    # Find the nearest cluster centers to the data points
    closest, _ = pairwise_distances_argmin_min(data, cluster_centers)
    new_labels = np.array([np.bincount(closest[labels == i]).argmax() for i in np.unique(labels)])
    return np.array([new_labels[label] for label in labels])

# Perform k-means clustering and relabel
n_clusters_range = range(2, 9)
best_kmeans, relabelled_labels = kmeans_and_relabel(MD_x_scaled, n_clusters_range, n_reps=10)

print("Best clustering labels after relabeling:")
print(relabelled_labels)
```

```python
def kmeans_inertia(data, n_clusters_range, n_reps):
    inertias = []
    for n_clusters in n_clusters_range:
        kmeans = KMeans(n_clusters=n_clusters, n_init=n_reps, random_state=1234)
        kmeans.fit(data)
        inertias.append(kmeans.inertia_)
    return inertias

# Perform k-means clustering and get inertia values
n_clusters_range = range(2, 9)
inertias = kmeans_inertia(MD_x_scaled, n_clusters_range, n_reps=10)

# Plot the inertia values
plt.figure(figsize=(10, 6))
plt.plot(n_clusters_range, inertias, marker='o', linestyle='-', color='b')
plt.title('K-Means Clustering Inertia vs. Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

```python
from sklearn.utils import resample

def bootstrap_kmeans(data, n_clusters_range, n_reps, n_boot):
    stability = {n_clusters: [] for n_clusters in n_clusters_range}
    
    for _ in range(n_boot):
        # Create a bootstrap sample
        boot_data = resample(data, random_state=1234)
        
        for n_clusters in n_clusters_range:
            kmeans = KMeans(n_clusters=n_clusters, n_init=n_reps, init='k-means++', random_state=1234)
            kmeans.fit(boot_data)
            stability[n_clusters].append(kmeans.labels_)
    
    # Compute stability metrics (e.g., cluster label agreement)
    cluster_stability = {n_clusters: 0 for n_clusters in n_clusters_range}
    
    for n_clusters in n_clusters_range:
        labels_list = stability[n_clusters]
        
        # Compute cluster label agreement
        consensus_matrix = np.zeros((n_clusters, n_clusters))
        
        for labels in labels_list:
            for i in range(len(labels)):
                for j in range(len(labels)):
                    if labels[i] == labels[j]:
                        consensus_matrix[labels[i], labels[j]] += 1
        
        # Normalize consensus matrix and compute mean similarity
        consensus_matrix /= len(labels_list)
        cluster_stability[n_clusters] = np.mean(np.diag(consensus_matrix))
    
    return cluster_stability

# Perform bootstrapping and get stability metrics
n_clusters_range = range(2, 9)
cluster_stability = bootstrap_kmeans(MD_x_scaled, n_clusters_range, n_reps=5, n_boot=20)

# Plot stability metrics
plt.figure(figsize=(10, 6))
plt.plot(cluster_stability.keys(), cluster_stability.values(), marker='o', linestyle='-', color='b')
plt.title('Cluster Stability vs. Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Stability Metric')
plt.grid(True)
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, n_init=10, init='k-means++', random_state=1234)
kmeans.fit(MD_x_scaled)
labels = kmeans.labels_

# Function to calculate cluster stability
def calculate_cluster_stability(data, labels):
    n_clusters = len(np.unique(labels))
    stability = np.zeros(n_clusters)
    
    for cluster_id in range(n_clusters):
        cluster_data = data[labels == cluster_id]
        other_data = data[labels != cluster_id]
        
        # Calculate stability for the current cluster
        distances = pairwise_distances_argmin_min(cluster_data, other_data)[1]
        stability[cluster_id] = np.mean(distances)
    
    return stability

# Calculate stability for the segments
stability = calculate_cluster_stability(MD_x_scaled, labels)

# Plot segment stability
plt.bar(range(4), stability, tick_label=[f'Segment {i}' for i in range(4)], color='b', edgecolor='black')
plt.xlabel('Segment Number')
plt.ylabel('Segment Stability')
plt.title('Segment Stability for 4 Clusters')
plt.ylim(0, 1)
plt.grid(True)
plt.show()
```

```python
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
def model_based_clustering(data, min_clusters, max_clusters, n_init=10):
    best_gmm = None
    best_n_clusters = None
    best_score = -np.inf
    silhouette_scores = []

    for n_clusters in range(min_clusters, max_clusters + 1):
        gmm = GaussianMixture(n_components=n_clusters, n_init=n_init, random_state=1234)
        gmm.fit(data)
        labels = gmm.predict(data)
        
        # Compute silhouette score for clustering evaluation
        score = silhouette_score(data, labels)
        silhouette_scores.append(score)
        
        if score > best_score:
            best_score = score
            best_gmm = gmm
            best_n_clusters = n_clusters
    
    return best_gmm, best_n_clusters, silhouette_scores

# Perform model-based clustering
min_clusters = 2
max_clusters = 8
best_gmm, best_n_clusters, silhouette_scores = model_based_clustering(MD_x_scaled, min_clusters, max_clusters)

# Print the best model and number of clusters
print(f"Best GMM Model: {best_gmm}")
print(f"Best Number of Clusters: {best_n_clusters}")

# Plot silhouette scores
plt.figure(figsize=(10, 6))
plt.plot(range(min_clusters, max_clusters + 1), silhouette_scores, marker='o', linestyle='-', color='b')
plt.title('Silhouette Score vs. Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()
```

```python
from scipy.spatial import ConvexHull

pca = PCA(n_components=2)
MD_x_pca = pca.fit_transform(MD_x)

# Perform KMeans clustering
kmeans = KMeans(n_clusters=4, random_state=1234).fit(MD_x)
clusters = kmeans.labels_

# Plot the results
plt.figure(figsize=(12, 8))
for i in range(4):
    cluster_points = MD_x_pca[clusters == i]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f'Cluster {i+1}', s=50)

    # Draw convex hull
    if len(cluster_points) > 2:  # Convex hull calculation needs at least 3 points
        hull = ConvexHull(cluster_points)
        for simplex in hull.simplices:
            plt.plot(cluster_points[simplex, 0], cluster_points[simplex, 1], 'k-')

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Clusters Projected onto the First Two Principal Components')
plt.legend()
plt.show()
```
